About the week: we have the MIR data from week 16~46, while Becky's team uses only 22 to 45.   
Here's their explanation in manusripts:   
The temporal window included in this study was the 24-week time-period from late May to late October (weeks 22 to 45), which corresponds to the timing of mosquito activity and human WNV illness, with data for the years from 2005 to 2016. The years from 2002 to 2004, during which Illinois had its first invasion from WNV, were excluded in this analysis because of the absence of mosquito testing data.

## Transfer weekly FID-MIR data to annual FID-MIR Data

```python
# Initialization
list_annual_mir_mean = []
for i in range(len(year_num.index)):
    dict_each_fid_mirmean[i] = {}
    list_annual_mir_mean.append(dict_each_fid_mirmean[i])
```

```python
# Transfer weekly FID-MIR data to annual FID-MIR data
# Took some time

for index,row in mir_file.iterrows():
    row_year = row['year']
    row_fid = row['INPUT_FID']
    row_mir = row['mirmean']
    
    i = 0
    for year in year_num:
        if row_year == year:
            list_annual_mir_mean[i][row_fid] = list_annual_mir_mean[i].get(row_fid, 0) + row_mir
            break
        i = i+1
        
```

```python
# write csv
mir_data_dir = "./MIR_data/"
year_basis = 2005
for i in range(12):
    with open(mir_data_dir + "Annual_FID_MIR_" + str(year_basis) + ".csv", "w") as fid_mir_csv_file:
        writer = csv.writer(fid_mir_csv_file)
        for key, value in list_annual_mir_mean[i].items():
            writer.writerow([key, value])
    year_basis = year_basis + 1
```

```python
# Read back
mir_data_dir = "./MIR_data/"
year_basis = 2005

for i in range(12):
    with open(mir_data_dir + "Annual_FID_MIR_" + str(year_basis) + ".csv") as fid_mir_csv_file:
        reader = csv.reader(fid_mir_csv_file)
        list_annual_mir_mean[i] = dict(reader)
    year_basis = year_basis + 1
```

## Transfer annual FID-MIR data to annual HEX-MIR data

```python
pop_hexagons_file
```

```python
# Initialization
list_annual_hex_mir = []
dict_each_hex_mir = {}
for i in range(len(year_num.index)):
    dict_each_hex_mir[i] = {}
    list_annual_hex_mir.append(dict_each_hex_mir[i])
```

```python
# Make a corresponding FID-HEXID dict

dict_fid_hexid = dict()
for index, row in pop_hexagons_file.iterrows():
    fid = row["Input_FID"]
    hexid = row["hexid"]
    dict_fid_hexid[fid] = hexid
```

```python
# Transfer annual FID-MIR data to annual HEX-MIR data

for i in range(12):
    annual_fid_mir = list_annual_mir_mean[i]
    for key, value in annual_fid_mir.items():
        hex_id = dict_fid_hexid.get(float(key))
        list_annual_hex_mir[i][hex_id] = value
```

```python
# write csv
mir_data_dir = "./MIR_data/"
year_basis = 2005
for i in range(12):
    with open(mir_data_dir + "Annual_HEXID_MIR_" + str(year_basis) + ".csv", "w") as hex_mir_csv_file:
        writer = csv.writer(hex_mir_csv_file)
        for key, value in list_annual_hex_mir[i].items():
            writer.writerow([key, value])
    year_basis = year_basis + 1
```

```python
# Read back
mir_data_dir = "./MIR_data/"
year_basis = 2005

for i in range(12):
    with open(mir_data_dir + "Annual_HEXID_MIR_" + str(year_basis) + ".csv") as hex_mir_csv_file:
        reader = csv.reader(hex_mir_csv_file)
        list_annual_hex_mir[i] = dict(reader)
    year_basis = year_basis + 1
```

```python
# Initialization
list_annual_tract_mir = []
dict_each_tract_mir = {}
for i in range(len(year_num.index)):
    dict_each_tract_mir[i] = {}
    list_annual_tract_mir.append(dict_each_tract_mir[i])
```

```python
float(list_annual_hex_mir[0].get("1756.0"))
```

```python
# Transfer annual HEX-MIR data to annual Tract-MIR data

for index, row in overlay_file.iterrows():
    hex_id = str(row["hexid"])
    tract_name_overlay = row["NAMELSAD10"][13:]
    hex_area = row["Shape__Area"]
    hex_in_tract_area = row["Area in Square Kilometers"] * 1000000
    ratio_hex_of_tract = hex_in_tract_area/hex_area
    
    for i in range(12):
        list_annual_tract_mir[i][tract_name_overlay] = (list_annual_tract_mir[i].get(tract_name_overlay,0) 
                                                        + ratio_hex_of_tract 
                                                        * float(list_annual_hex_mir[i].get(hex_id)))
```

```python
# write csv
mir_data_dir = "./MIR_data/"
year_basis = 2005
for i in range(12):
    with open(mir_data_dir + "Annual_TRACT_MIR_" + str(year_basis) + ".csv", "w") as tract_mir_csv_file:
        writer = csv.writer(tract_mir_csv_file)
        for key, value in list_annual_tract_mir[i].items():
            writer.writerow([key, value])
    year_basis = year_basis + 1
```

```python
# Read back
mir_data_dir = "./MIR_data/"
year_basis = 2005

for i in range(12):
    with open(mir_data_dir + "Annual_TRACT_MIR_" + str(year_basis) + ".csv") as tract_mir_csv_file:
        reader = csv.reader(tract_mir_csv_file)
        list_annual_tract_mir[i] = dict(reader)
    year_basis = year_basis + 1
```
